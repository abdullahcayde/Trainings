```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('planets')
df.head()                      
```

```python
df.describe().T
```

## Groupby

```python
df.groupby('method')['orbital_period'].mean()
```

```python
df.groupby('number').describe().T
```

```python
df.groupby('number').mean()
```

```python
df.groupby('number').count()
```

```python
df.groupby('method')['mass'].mean()
```

```python
df.groupby('method')['mass'].describe()
```

## Aggregate

```python
df.groupby('method')['orbital_period'].aggregate(['min', max, np.median])
```

```python
df.groupby('method')[['orbital_period', 'distance']].aggregate(['count', 'min', max, 'mean', 'std'])
```

```python
df.groupby('method').aggregate({'orbital_period' : [min, max],
                               'distance' : ['mean', np.median]})
```
