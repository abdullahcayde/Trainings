```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.DataFrame({'gruplar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'degisken1': [10,23,33,22,11,99],
                   'degisken2': [100,253,333,262,111,969]},
                   columns = ['gruplar', 'degisken1', 'degisken2'])
df
```

# Filter

```python
def fonk1(x):
    return x['degisken1'].std() > 9
```

```python
df.groupby('gruplar').filter(fonk1)
```

```python
df.groupby('gruplar').std()
```

```python
df.groupby('gruplar').max()
```

```python
def fonk2(x):
    return (x['degisken1'].max() < x['degisken2'].max()) & (x['degisken2'].max() > 260)
```

```python
df.groupby('gruplar').filter(fonk2)
```
