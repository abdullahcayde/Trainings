```python
import pandas as pd
import numpy as np

import seaborn as sns
sns.set() #setting the default seaborn style for our plots
sns.set_theme(style="whitegrid")

pd.set_option('display.max_columns', None)

import warnings
warnings.simplefilter(action='ignore')
```

```python
pergamo = pd.read_csv('pergamo.csv')
df = pergamo.copy()
df.rename(columns= lambda x: x.replace(' ', '_').replace('#', '').lower(), inplace=True)
df.head(2)
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.order_date = pd.to_datetime(df.order_date)
df[['day', 'month', 'year']] = df['order_date'].apply(lambda x: pd.Series([x.day, x.month, x.year]))
df[['day', 'month', 'year']].head(2)
```

```python
df.describe().T
```

```python
writer = pd.ExcelWriter('market_abd.xlsx')

# df.info(), df.describe(), df.quantile()
df.dtypes.reset_index(name='Dtype').rename(columns={'index' : 'Column'}).to_excel(writer, 'info')
df.describe(include='all').T.to_excel(writer, 'describe')
df.quantile([0.01,0.25,0.50,0.75,0.99]).T.to_excel(writer, 'quantile')

writer.save()
```
