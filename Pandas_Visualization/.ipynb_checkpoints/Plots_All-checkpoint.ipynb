```python
import pandas as pd 
import numpy as np
import seaborn as sns
sns.set_theme(style="whitegrid")
import matplotlib.pyplot as plt

import warnings
warnings.simplefilter(action='ignore')
```

```python
path = '/Users/macbook/Desktop/projects/Github_Repositories/Redi'
immo = pd.read_csv(f'{path}/immo_original.csv', usecols=[1,2,3,4,5,6])
df = immo.copy()
df['Age'] = 2022 - df.Year_Construction
df.head()
```

# sns.relpot() [Relation Plot]

```python
g = sns.relplot(x=, y=, hue=, data=,
            
            
            style='column name' ,#Nokta Sekillerini Degistirir
            palette=[, , , ,] ,# Istedigin renkleri atarsin) 
            markers=['X', 'o'],
            s=100 # Size of markers
            
            col= 'column name',
            col_wrap=3  ,# Fazla Plot varsa 3 3 gosterir
           
            row='column name',
            ci=None)

g.set_xlabels('')
g.set_ylabels('')

# palette = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
```

```python
sns.relplot(x='Region', y='Age', data=df);
plt.xticks(rotation=90);
```

```python
sns.relplot(x='Region', y='Age', hue='Age', data=df);
plt.xticks(rotation=90);
```

```python
#kind = 'line' ===> mean of Age with Line Plot ci=0.95
sns.relplot(x='Region', y='Age', data=df, kind='line');
plt.xticks(rotation=90);
```

```python
plt.figure(figsize=(5,5))
sns.relplot(x='Region', y='Rent', data=df.loc[df.Age < 5],
           col= 'Age')
plt.xticks(rotation=90);
```

```python
plt.figure(figsize=(3,3))
sns.relplot(x='Region', y='Rent', data=df.loc[df.Age < 6],
           col= 'Age', kind='line')
plt.xticks(rotation=90);
```

```python
sns.relplot(x='Region', y='Rent', data=df.loc[df.Age < 4],
           row= 'Age', kind='line')
plt.xticks(rotation=90);
```
